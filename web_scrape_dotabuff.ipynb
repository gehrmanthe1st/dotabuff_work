```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.dotabuff.com/heroes/winning?date=month')
```

```python
results = []
content = driver.page_source
soup = BeautifulSoup(content, 'lxml')
```

```python
soup
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
name_tags = soup.find_all('a', class_ = 'link-type-hero')
name_tags
```

```python
char_names = []
for name in name_tags:
    char_names.append(name.text)

char_names
```

```python
import csv
```

```python
csv_filename = 'character_trend.csv'

with open(csv_filename, 'w', newline = '') as csvfile:
    csv_writer = csv.writer(csvfile)
    for char in char_names:
        csv_writer.writerow([char])

csvfile.close()
```

```python
with open(csv_filename, 'r') as csvfile:
    csv_reader = csv.reader(csvfile)
    data = list(csv_reader)

new_row = ['Names']
data.insert(0, new_row)
data.insert(1, [])

with open(csv_filename, 'w', newline = '') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerows(data)

csvfile.close()
```
